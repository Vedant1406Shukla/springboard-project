```python
import google.generativeai as genai

# -----------------------------------
# Configure Gemini API
# -----------------------------------
genai.configure(api_key="AIzaSyAa67mw7Mqj3iBTZVyxm_L2djS-Mz1V_VA")

model = genai.GenerativeModel("gemini-2.5-flash")


def ask_gemini(prompt):
    """Send a prompt to Gemini and return the text output."""
    response = model.generate_content(prompt)
    return response.text


def generate_subquestions(topic):
    prompt = f"""
    Generate exactly 3 analytical sub-questions related to the research topic:
    "{topic}"

    Return only a numbered list (1, 2, 3).
    """
    return ask_gemini(prompt)


def answer_subquestions(subquestions):
    answers = []
    for q in subquestions.split("\n"):
        if q.strip():
            prompt = f"Give a concise 3‚Äì4 sentence answer for the question: {q}"
            answers.append(ask_gemini(prompt))
    return answers


def summarise_results(topic, answers):
    combined = "\n".join(answers)
    prompt = f"""
    Summarise the following insights into a single cohesive paragraph on:
    "{topic}"

    Insights:
    {combined}
    """
    return ask_gemini(prompt)


# -----------------------------------
# MAIN PIPELINE
# -----------------------------------
if __name__ == "__main__":
    topic = input("Enter research topic: ")

    print("\nüîç Generating sub-questions...\n")
    sub_qs = generate_subquestions(topic)
    print(sub_qs)

    print("\nüìù Getting short answers...\n")
    answers = answer_subquestions(sub_qs)
    for a in answers:
        print("- " + a + "\n")

    print("\nüìò Final Summary:\n")
    summary = summarise_results(topic, answers)
    print(summary)
```
